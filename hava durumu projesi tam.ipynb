SAYFAYA İSTEK ATMA FONKSİYONUN OLUŞTURULMASI

```python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Mozilla/5.0"})
    soup = bts(result.text, "html.parser")
    return soup
```

20. 7 Günlük Tahminlerin DataFrame Olarak Elde Edilmesi

```python
features_morning = []

for day in days[:7]:
    
    html = getAndParseURL(day)
    
    try:
        day_name = html.find("div", {"class":"subnav-pagination"}).text.split(",")[0].strip()
    except:
        day_name = np.nan
    try:
        date = html.find("div", {"class":"subnav-pagination"}).text.split(",")[1].strip()
    except:
        date = np.nan
    try:
        morning = html.find("h2", {"class":"title"}).text
    except:
        morning = np.nan
    try:
        morning_temp = html.find("div", {"class":"temperature"}).find(text=re.compile("°")).strip()
    except:
        morning_temp = np.nan
    try:
        morning_uv = html.find("div", {"class":"panels"}).find(text="Maks UV İndeksi").findNext().text
    except:
        morning_uv = np.nan
    try:
        morning_wind = html.find("div", {"class":"panels"}).find(text="Rüzgar").findNext().text
    except:
        morning_wind = np.nan
    try:
        morning_rain_exp = html.find("div", {"class":"panels"}).find(text="Yağış Beklentisi").findNext().text
    except:
        morning_rain_exp = np.nan
    try:
        morning_downpour = html.find("div", {"class":"panels"}).find(text="Gök Gürültülü Sağanak Yağış Olasılığı").findNext().text
    except:
        morning_downpour = np.nan
    try:
        morning_rain = html.find("div", {"class":"panels"}).find(text="Yağış").findNext().text
    except:
        morning_rain = np.nan
    try:
        morning_cloud = html.find("div", {"class":"panels"}).find(text="Bulutlarla Kaplı").findNext().text
    except:
        morning_cloud = np.nan
        
    features_morning.append([date, day_name, morning, morning_temp, morning_wind, morning_rain_exp,
                             morning_downpour, morning_rain, morning_cloud, morning_uv])
    
df_morning_columns = ["Tarih", "Gün", "Gün Dönemi", "Sıcaklık", "Rüzgar", "Yağış Beklentisi",
                     "Gök Gürültülü Sağanak Yağış Olasılığı", "Yağış", "Bulutlarla Kaplı", "Maks UV İndeksi"]

df_morning = pd.DataFrame.from_records(features_morning, columns=df_morning_columns)
df_morning
```

```python
features_evening = []

for day in days[:7]:
    
    html = getAndParseURL(day)
    
    try:
        day_name = html.find("div", {"class":"subnav-pagination"}).text.split(",")[0].strip()
    except:
        day_name = np.nan
    try:
        date = html.find("div", {"class":"subnav-pagination"}).text.split(",")[1].strip()
    except:
        date = np.nan
    try:
        evening = html.find("div", {"class":"page-content content-module"}).findAll("h2", {"class":"title"})[1].text
    except:
        evening = np.nan
    try:
        evening_temp = html.find("div", {"class":"page-content content-module"}).findAll("div", {"class":"temperature"})[1].find(text=re.compile("°")).strip()
    except:
        evening_temp = np.nan
    try:
        evening_wind = html.find("div", {"class":"page-content content-module"}).findAll("div", {"class":"panels"})[1].find(text="Rüzgar").findNext().text
    except:
        evening_wind = np.nan
    try:
        evening_rain_exp = html.find("div", {"class":"page-content content-module"}).findAll("div", {"class":"panels"})[1].find(text="Yağış Beklentisi").findNext().text
    except:
        evening_rain_exp = np.nan
    try:
        evening_downpour = html.find("div", {"class":"page-content content-module"}).findAll("div", {"class":"panels"})[1].find(text="Gök Gürültülü Sağanak Yağış Olasılığı").findNext().text
    except:
        evening_downpour = np.nan
    try:
        evening_rain = html.find("div", {"class":"page-content content-module"}).findAll("div", {"class":"panels"})[1].find(text="Yağış").findNext().text
    except:
        evening_rain = np.nan
    try:
        evening_cloud = html.find("div", {"class":"page-content content-module"}).findAll("div", {"class":"panels"})[1].find(text="Bulutlarla Kaplı").findNext().text
    except:
        evening_cloud = np.nan
        
    features_evening.append([date, day_name, evening, evening_temp, evening_wind,
                             evening_rain_exp, evening_downpour, evening_rain, evening_cloud])
    
    
df_evening_columns = ["Tarih", "Gün", "Gün Dönemi", "Sıcaklık", "Rüzgar", "Yağış Beklentisi",
                      "Gök Gürültülü Sağanak Yağış Olasılığı", "Yağış", "Bulutlarla Kaplı"]

df_evening = pd.DataFrame.from_records(features_evening, columns=df_evening_columns)
df_evening
```

```python
df = pd.concat([df_morning, df_evening])
df.sort_values(["Tarih", "Gün Dönemi"], ascending=[True, False], inplace=True)
df
```
